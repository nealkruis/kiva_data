```python
def createResultsDict(dir):
    dict = {}
    dict['dir'] = dir
    dict['df'] = pd.read_csv(pdir + '/' + dir + '/Timeseries.csv',
                                     header=0,
                                     names=['time','total','foundation','wall_total','wall','floor_total','floor'],
                                     usecols=[0,1,2,3,4,5,6],
                                     parse_dates=True,
                                     index_col=0,
                                     na_values=['null'])

    dict['warnings'] = False
    with open(pdir + '/' + dir + '/log.out') as f:
        lines = f.readlines()
    for line in lines:
        if 'Warning:' in line:
            dict['warnings'] = True
        if 'Elapsed Time:' in line:
            time_string = re.search('Elapsed Time: (.*)\n',line).group(1)
            continue
    time = datetime.strptime(time_string,"%H:%M:%S.%f")
    delta = timedelta(hours=time.hour, minutes=time.minute, seconds=time.second, microseconds=time.microsecond)
    dict['time'] = delta
    
    dict['S'] = dir[1]
    dict['AP'] = dir[4:6]
    dict['F'] = dir[7]
    dict['C'] = dir[8:10]
    dict['I'] = dir[11:13]
    if len(dir) > 13:
        dict['R'] = dir[14:16]
    else:
        dict['R'] = '3D'
    dict['total'] = sum(list(dict['df'].total))/1000.0
    dict['foundation'] = sum(list(dict['df'].foundation))/1000.0
    dict['floor'] = sum(list(dict['df'].floor))/1000.0
    dict['floor_total'] = sum(list(dict['df'].floor_total))/1000.0
    dict['floor_area'] = dict['floor_total']/dict['floor']
    if dict['F'] != 'S':
        dict['wall'] = sum(list(dict['df'].wall))/1000.0
        dict['wall_total'] = sum(list(dict['df'].wall_total))/1000.0
        dict['wall_area'] = dict['wall_total']/dict['wall'] 
    else:
        dict['wall'] = 0.0
        dict['wall_total'] = 0.0
        dict['wall_area'] = 0.0 


     
    return dict
```

```python
results = {}
#dfs = []
dirs = []
for dir in os.walk(pdir).next()[1]:
    if dir[0] == 'S':
        dirs.append(dir)
res = v.map(createResultsDict,dirs)

#dfs.append(results[dir]['df'])
#df = pd.concat(results,axis=1)
while not res.ready():
    time.sleep(5)
    print res.progress/float(len(dirs))*100

for r in res:
    results[r['dir']] = r
    if r['warnings']:
        print "Warnings in: " + r['dir']
```

```python
def createPlot(shape):
    fig = plt.figure()
    plot_data = {}
    
    gs = gridspec.GridSpec(3, 3, width_ratios=[10,6,6])
    
    shape_name = {'S':'Square',
                  '#':'Narrow-H',
                  'H':'Wide-H',
                  '+':'Cross'}

    APRatios = {'05':5,
                '15':15,
                '45':45}
    
    Foundations = {'B':'Basement',
                   'C':'Crawlspace',
                   'S':'Slab'}
    
    Basements = {'UI00':'Uninsulated',
                 'EH05':'R-5 Half-Wall Ext.', 
                 'EF05':'R-5 Full-Wall Ext.', 
                 'IF05':'R-5 Full-Wall Int.', 
                 'EH20':'R-20 Half-Wall Ext.',
                 'EF20':'R-20 Full-Wall Ext.',
                 'IF20':'R-20 Full-Wall Int.'}
    border = []
    for o in ['UI00', 
              'EH05', 
              'EF05', 
              'IF05', 
              'EH20', 
              'EF20', 
              'IF20']:
        border.append(Basements[o])

    Crawlspaces = {'UI00':'Uninsulated',
                   'EV05':'R-5 Ext. Vert.',
                   'IV05':'R-5 Int. Vert.',
                   'IB05':'R-5 Int. Vert. & Hor.',
                   'EV20':'R-20 Ext. Vert.',
                   'IV20':'R-20 Int. Vert.',
                   'IB20':'R-20 Int. Vert. & Hor.'}
    corder = []
    for o in ['UI00', 
              'EV05', 
              'IV05', 
              'IB05', 
              'EV20', 
              'IV20', 
              'IB20']:
        corder.append(Crawlspaces[o])

    Slabs = {'UI00':'Uninsulated',
             'EH05':'R-5 Ext. Hor.',
             'EV05':'R-5 Ext. Vert.',
             'IV05':'R-5 Int. Vert.',
             'IH05':'R-5 Int. Hor.',
             'IW05':'R-5 Int. Whole-Slab',
             'EH20':'R-20 Ext. Hor.',
             'EV20':'R-20 Ext. Vert.',
             'IV20':'R-20 Int. Vert.',
             'IH20':'R-20 Int. Hor.',
             'IW20':'R-20 Int. Whole-Slab'}
    sorder = []
    for o in ['UI00',
              'EH05',
              'EV05', 
              'IV05',
              'IH05',
              'IW05',
              'EH20',
              'EV20',
              'IV20',
              'IH20',
              'IW20']:
        sorder.append(Slabs[o])

    strat = ['AC',
             'AL',
             'NC',
             'NL',
             'PC',
             'BC',
             'RR']

    colors = {'AC': palette[0],
              'AL': palette[1],
              'NC': palette[2],
              'NL': palette[3],
              'PC': palette[4],
              'BC': palette[5],
              'RR': palette[6]}
    
    markers = {'AC': 'o',
               'AL': 'd',
               'NC': 'o',
               'NL': 'd',
               'PC': 'o',
               'BC': 'o',
               'RR': 's'}

    markerfills = {'AC': colors['AC'],
                   'AL': colors['AL'],
                   'NC': 'w',
                   'NL': 'w',
                   'PC': 'w',
                   'BC': colors['BC'],
                   'RR': colors['RR']}

    markeredges = colors
    
    linestyles = {'AC': '-',
                  'AL': '-',
                  'NC': '--',
                  'NL': '--',
                  'PC': ':',
                  'BC': '--',
                  'RR': '-'}
    
    linewidths = {'AC': 2,
                  'AL': 2,
                  'NC': 2,
                  'NL': 2,
                  'PC': 2,
                  'BC': 2,
                  'RR': 2}
    
    labels = {'AC': 'A/P: Cylindrical',
              'AL': 'A/P: Cartesian',
              'NC': 'A/P & (A/P)Neg: Cylindrical',
              'NL': 'A/P & (A/P)Neg: Cartesian',
              'PC': 'A/P & PNeg: Cylindrical',
              'BC': 'A & P: Cylindrical',
              'RR': 'Rounded Rectangle (A & P)'}

    for a in APRatios:
        for f in Foundations:
            plot_data[a+f] = {}
            for st in strat:
                sh = shape
                if st == 'AC':
                    sh = 'C'
                    suffix = ''
                elif st == 'AL':
                    sh = 'L'
                    suffix = ''
                elif st == 'NC':
                    suffix = 'RNC'
                elif st == 'NL':
                    suffix = 'RNL'
                elif st == 'PC':
                    suffix = 'RPC'
                elif st == 'BC':
                    suffix = 'RBC'
                
                plot_data[a+f][st] = {}
                if f == 'B':
                    for b in Basements:
                        k_3D = 'S'+shape+'AP'+a+'F'+f+b[0:2]+'I'+b[2:]
                        res_3D = results[k_3D]['total']/1000
                        if st == 'RR':
                            k_c = 'S'+sh+'AP'+a+'F'+f+b[0:2]+'I'+b[2:]+'RRC'
                            k_l = 'S'+sh+'AP'+a+'F'+f+b[0:2]+'I'+b[2:]+'RRL'
                            res_k = (results[k_c]['wall_total']+results[k_l]['wall_total']+results[k_c]['floor_total']+results[k_l]['floor_total'])/1000
                        else:
                            k = 'S'+sh+'AP'+a+'F'+f+b[0:2]+'I'+b[2:]+suffix
                            res_k = (results[k]['wall']*results[k_3D]['wall_area']+results[k]['floor']*results[k_3D]['floor_area'])/1000
                        plot_data[a+f][st][Basements[b]] = (res_k - res_3D)/res_3D
                if f == 'C':
                    for b in Crawlspaces:
                        k_3D = 'S'+shape+'AP'+a+'F'+f+b[0:2]+'I'+b[2:]
                        res_3D = results[k_3D]['total']/1000
                        if st == 'RR':
                            k_c = 'S'+sh+'AP'+a+'F'+f+b[0:2]+'I'+b[2:]+'RRC'
                            k_l = 'S'+sh+'AP'+a+'F'+f+b[0:2]+'I'+b[2:]+'RRL'
                            res_k = (results[k_c]['wall_total']+results[k_l]['wall_total']+results[k_c]['floor_total']+results[k_l]['floor_total'])/1000
                        else:
                            k = 'S'+sh+'AP'+a+'F'+f+b[0:2]+'I'+b[2:]+suffix
                            res_k = (results[k]['wall']*results[k_3D]['wall_area']+results[k]['floor']*results[k_3D]['floor_area'])/1000
                        plot_data[a+f][st][Crawlspaces[b]] = (res_k - res_3D)/res_3D
                if f == 'S':
                    for b in Slabs:
                        k_3D = 'S'+shape+'AP'+a+'F'+f+b[0:2]+'I'+b[2:]
                        res_3D = results[k_3D]['total']/1000
                        if st == 'RR':
                            k_c = 'S'+sh+'AP'+a+'F'+f+b[0:2]+'I'+b[2:]+'RRC'
                            k_l = 'S'+sh+'AP'+a+'F'+f+b[0:2]+'I'+b[2:]+'RRL'
                            res_k = (results[k_c]['wall_total']+results[k_l]['wall_total']+results[k_c]['floor_total']+results[k_l]['floor_total'])/1000
                        else:
                            k = 'S'+sh+'AP'+a+'F'+f+b[0:2]+'I'+b[2:]+suffix
                            plot_data[a+f][st][Slabs[b]] = (results[k]['wall']*results[k_3D]['wall_area']+results[k]['floor']*results[k_3D]['floor_area'])/1000
                            res_k = (results[k]['wall']*results[k_3D]['wall_area']+results[k]['floor']*results[k_3D]['floor_area'])/1000
                        plot_data[a+f][st][Slabs[b]] = (res_k - res_3D)/res_3D
    
    ax = {} 
    for p in plot_data:

        if p[2] == 'B':
            order = border
            col = 3
        elif p[2] == 'C':
            order = corder
            col = 2
        elif p[2] == 'S':
            order = sorder
            col = 1

        if p[0:2] == '05':
            row = 1
            m = 50
        elif p[0:2] == '15':
            row = 2    
            m = 25
        elif p[0:2] == '45':
            row = 3
            m = 10

        plot_num = col + 3*(row-1)
        ax[p] = fig.add_subplot(gs[row-1,col-1])
        ax[p].axhline(0, color='k')
        for st in strat:
            values = OrderedDict(sorted(plot_data[p][st].items(),key=lambda i:order.index(i[0]))).values()
            keys = OrderedDict(sorted(plot_data[p][st].items(),key=lambda i:order.index(i[0]))).keys()
            ax[p].plot(values,label=labels[st], marker=markers[st], color=colors[st], linestyle=linestyles[st],linewidth=linewidths[st],markerfacecolor=markerfills[st],markeredgecolor=markeredges[st],markeredgewidth=1.0)
            ax[p].set_xticks(range(len(keys)))
            if row == 3:
                ax[p].set_xticklabels(keys, rotation=90)
            else:
                ax[p].set_xticklabels([])

        #ax[p].set_xlim(-0.5,len(keys))
        ax[p].set_ylim(-0.2,0.2)
        if row == 1 and col == 3:
            handles, labs = ax[p].get_legend_handles_labels()
            fig.legend(handles,labs,loc='upper center',ncol=3)
        if row == 1:
            ax[p].set_title(Foundations[p[2]])
        if col == 1:
            ax[p].set_ylabel('A / P = ' + str(round(float(p[0:2])*0.3048,1)) + ' m')
            ax[p].yaxis.set_major_formatter(FuncFormatter(to_percent))
        else:
            ax[p].set_yticklabels([])
            
    fig.subplots_adjust(left=None, bottom=None, right=None, top=0.9, wspace=0.15, hspace=0.15)
    fig.savefig('/Users/nkruis/Documents/workspace/Thesis/images/Approx-'+climate+'-'+shape_name[shape]+'.pdf')
    
```
